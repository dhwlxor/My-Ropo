```python
!pip install ipycanvas ipywidgets

import numpy as np
import tensorflow as tf
from ipycanvas import Canvas
from ipywidgets import Button, VBox, Output
from IPython.display import display
import matplotlib.pyplot as plt

from google.colab import files
uploaded = files.upload()  # mnist_cnn_model.h5 ì„ íƒ

model = tf.keras.models.load_model('mnist_cnn_model.h5')
print("âœ… ëª¨ë¸ ë¡œë“œ ì™„ë£Œ")

from ipycanvas import Canvas
from ipywidgets import Button, VBox, Output
from IPython.display import display
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

# ìº”ë²„ìŠ¤ ë° ë²„íŠ¼ UI ìƒì„±
canvas = Canvas(width=280, height=280, sync_image_data=True)
predict_btn = Button(description="ğŸ§  Predict")
clear_btn = Button(description="ğŸ§¹ Clear")
output = Output()

# ìˆ«ì ì¸ì‹ í•¨ìˆ˜
def recognize_digit():
    data = canvas.get_image_data(0, 0, 280, 280)
    img = np.array(data.data)[::4].reshape((280, 280))  # R ì±„ë„ë§Œ ì¶”ì¶œ
    img = 1 - (img / 255.0)                             # ìƒ‰ ë°˜ì „ ë° ì •ê·œí™”
    img = tf.image.resize(img[..., np.newaxis], (28, 28))  # 28x28 í¬ê¸°ë¡œ
    img = img.numpy().reshape(1, 28, 28, 1)             # ë°°ì¹˜ ì°¨ì› ì¶”ê°€

    pred = model.predict(img)
    pred_label = np.argmax(pred)

    plt.imshow(img.reshape(28, 28), cmap='gray')
    plt.title(f"ğŸ“¢ ì˜ˆì¸¡ ìˆ«ì: {pred_label}")
    plt.axis('off')
    plt.show()

# ë²„íŠ¼ í´ë¦­ ì´ë²¤íŠ¸
def on_predict_clicked(b):
    output.clear_output()
    with output:
        recognize_digit()

def on_clear_clicked(b):
    canvas.clear()

predict_btn.on_click(on_predict_clicked)
clear_btn.on_click(on_clear_clicked)

# UI ì¶œë ¥
display(VBox([canvas, VBox([predict_btn, clear_btn]), output]))

from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# ì´ë¯¸ì§€ ë¡œë”© ë° ì „ì²˜ë¦¬
img = Image.open('x.png').convert('L')  # í‘ë°± ë³€í™˜
img = img.resize((28, 28))              # MNIST í¬ê¸° ë§ì¶¤
img_array = np.array(img)

# ì´ë¯¸ì§€ ë°˜ì „ ë° ì •ê·œí™”
img_array = 255 - img_array             # í° ë°”íƒ•, ê²€ì€ ê¸€ì”¨ë¡œ
img_array = img_array / 255.0           # ì •ê·œí™”
img_array = img_array.reshape(1, 28, 28, 1)  # CNN ì…ë ¥ í˜•íƒœ

# ì˜ˆì¸¡
pred = model.predict(img_array)
pred_label = np.argmax(pred)

# ê²°ê³¼ ì¶œë ¥
plt.imshow(img, cmap='gray')
plt.axis('off')
plt.title(f'ğŸ“¢ ì˜ˆì¸¡ ìˆ«ì: {pred_label}')
plt.show()

```
